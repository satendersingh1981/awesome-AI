```python
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown,display
system_message="""Professional Summary
Technology leader and Enterprise Architect with 14+ years of experience in cloud data platforms, big data
engineering, and enterprise analytics. Proven track record in designing scalable, reliable, and cost-efficient
data architectures. Led initiatives that reduced Snowflake infrastructure costs by ~60% (~$2M/year) and AWS
infrastructure by ~$1M annually. Expertise in Snowflake, AWS, Databricks, Apache Airflow, and FinOps.
Passionate about integrating Generative AI and LLMs into enterprise data workflows."""
messages=[
        {"role":"system","content":system_message},
        {"role":"user","content":"what all technologies he knows?"}
]
load_dotenv()
client=OpenAI()
response = client.chat.completions.create(messages=messages,model="gpt-5.4-nano")
display(Markdown(response.choices[0].message.content))
```
