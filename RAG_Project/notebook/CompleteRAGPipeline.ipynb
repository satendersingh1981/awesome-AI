```python
import numpy as np # To convert list of vectors to numpy array
from sentence_transformers import SentenceTransformer # To generate embeddings for the document chunks
import chromadb # To create vector database and store the document chunks
from chromadb.config import Settings # To configure the ChromaDB client
import uuid # To generate unique IDs for the document chunks
from typing import List, Dict, Any, Tuple # For type annotations
from sklearn.metrics.pairwise import cosine_similarity # To calculate cosine similarity between vectors
```

```python
class EmbeddingManager:
    # handles document embedding generation usig sentence transformer
    def __init__(self,model_name: str = "BAAI/bge-base-en-v1.5"):  #all-MiniLM-L6-v2
        self.model_name=model_name
        self.model = None
        self.__load_model()

    def __load_model(self):
        try:
            print(f"🔄 Loading embedding model: {self.model_name}...")
            self.model =SentenceTransformer(self.model_name)
            print(f"✅ Model loaded successfully!, Dimension in model are :{self.model.get_embedding_dimension()}")
        except Exception as e:
            print(f"❌ Error loading model: {e}")
            raise e
    def generate_embeddings(self,texts: List[str]) -> np.ndarray: 
        # Generate embeddings for a list of chunks of text and return as numpy array
        try:
            if not self.model:
                raise ValueError("Model not loaded.Please load the model before generating embeddings.")
            
            print(f"🔄 Generating embeddings for {len(texts)} texts...")
            embeddings = self.model.encode(texts, show_progress_bar=True)
            return np.array(embeddings)
        except Exception as e:
            print(f"❌ Error generating embeddings: {e}")
            raise e
    
```

### Vector Store Class

```python
class VectorStore:
    # manages the vector database using ChromaDB to store document chunks and their embeddings
    """
        Initialize the vector store
        
        Args:
            collection_name: Name of the ChromaDB collection
            persist_directory: Directory to persist the vector store on local drive
    """
    def __init__(self, collection_name: str = "pdf_documents", persist_directory: str = "../data/vector_db"):
        self.collection_name = collection_name
        self.persist_directory = persist_directory
        self.client = None
        self.collection = None
        self.__initialize_vector_store()

    def __initialize_vector_store(self):
        # Initialize the ChromaDB client and collection
        try:
            # Create persistent ChromaDB client
            os.makedirs(self.persist_directory, exist_ok=True)
            
            print(f"🔄 Initializing ChromaDB vector store with collection name: {self.collection_name}...")
            self.client = chromadb.PersistentClient(path=self.persist_directory)
            self.collection = self.client.get_or_create_collection(
                name=self.collection_name,
                metadata={"description": "PDF document embeddings for RAG"})
            
            print(f"Vector store initialized. Collection: {self.collection_name}")
            print(f"Existing documents in collection: {self.collection.count()}")
        except Exception as e:
            print(f"❌ Error initializing vector store: {e}")
            raise e

    def reset_collection(self):
        """Delete and recreate the collection so stale persisted data can be rebuilt."""
        try:
            if self.client:
                self.client.delete_collection(name=self.collection_name)
                self.collection = self.client.get_or_create_collection(name=self.collection_name)
                print(f"✅ Collection '{self.collection_name}' reset successfully.")
        except Exception as e:
            print(f"❌ Error resetting collection: {e}")
            raise e
    def add_documents(self,documents: List[Document],embeddings: np.ndarray):

        # Add documents and their embeddings to the vector store
        """
        Add documents and their embeddings to the vector store
        
        Args:
            documents: List of LangChain documents
            embeddings: Corresponding embeddings for the documents
        """
    
        if not self.collection:
            raise ValueError("Vector store not initialized. Please initialize the vector store before adding documents.")
        
        if len(documents) != len(embeddings):
            raise ValueError("Number of documents must match number of embeddings")

        print(f"🔄 Adding {len(documents)} documents to the vector store...")

        # Prepare data for ChromaDB
        ids = []
        metadatas = []
        documents_text = []
        embeddings_list = []
        for i,(doc,embedding) in enumerate(zip(documents,embeddings)):
            # Generate unique ID for each document chunk
            doc_id = f"doc_{uuid.uuid4().hex[:8]}_{i}"
            ids.append(doc_id)

            # Prepare metadata
            metadata = dict(doc.metadata)
            metadata['source'] = metadata.get('source', '')
            metadata['page'] = metadata.get('page', '')
            metadata['doc_index'] = i
            metadata['content_length'] = len(doc.page_content)
            metadatas.append(metadata)

            # Store document text and embedding
            documents_text.append(doc.page_content)
            embeddings_list.append(embedding.tolist())
        

        # Add to collection
        try:
            self.collection.add(
                ids=ids,
                embeddings=embeddings_list,
                metadatas=metadatas,
                documents=documents_text
            )
            print(f"Successfully added {len(documents)} documents to vector store")
            print(f"Total documents in collection: {self.collection.count()}")
            
        except Exception as e:
            print(f"Error adding documents to vector store: {e}")
            raise
```

### Augumented Reterival pipeline

```python
class RAGRetriever:
    # handles the retrieval of relevant data chunks from the vector store based on user query
    def __init__(self,vector_store,embedding_manager):
        self.vector_store = vector_store
        self.embedding_manager = embedding_manager

    def retrieve(self,query:str,top_k :int =5 ,similarity_threshold: float=0.0) -> List[Dict[str,Any]]:
        # Retrieve relevant chunks from the vector store based on user query
        try:
            # Generate embedding for the user query
            print(f"🔄 Generating embedding for the query: '{query}'...")
            query_embedding = self.embedding_manager.generate_embeddings([query])[0]

            # Retrieve documents from the vector store
            print(f"🔄 Retrieving documents from the vector store...")
            try:
                results = self.vector_store.collection.query(
                    query_embeddings=query_embedding.tolist(),
                    n_results=max(top_k * 10, 20),
                    include=['documents', 'metadatas', 'distances']
                )

                retrieved_docs = []
                documents = results.get('documents', [])[0] if results.get('documents') else []
                metadatas = results.get('metadatas', [])[0] if results.get('metadatas') else []
                distances = results.get('distances', [])[0] if results.get('distances') else []
                ids = results.get('ids', [])[0] if results.get('ids') else [None] * len(documents)

                for i,(doc_id,document,metadata,distance) in enumerate(zip(ids,documents,metadatas,distances)):
                    similarity_score = 1 - distance
                    if similarity_score >= similarity_threshold:
                        retrieved_docs.append({
                            "id" : doc_id,
                            "content": document,
                            "metadata": metadata,
                            "similarity_score": similarity_score,
                            'distance': distance,
                            'rank': i + 1
                        })
                    if len(retrieved_docs) >= top_k:
                        break

                print(f"✅ Retrieved {len(retrieved_docs)} relevant documents for the query.")
                return retrieved_docs
            except Exception as e:
                print(f"❌ Error querying vector store: {e}")
                raise e 
        except Exception as e:
            print(f"❌ Error in retrieval process: {e}")
            raise e
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.messages import HumanMessage, SystemMessage
from openai import OpenAI
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
class OpenAILLM:
    # handles interaction with OpenAI LLM to generate responses based on retrieved documents and user query
    def __init__(self,model_name:str="gpt-5.4-nano",temperature:float=0.7, api_key: str =None):
        self.model_name=model_name
        self.temperature=temperature
        self.api_key = api_key
        if not self.api_key or os.getenv("OPENAI_API_KEY") is None:
            raise ValueError("OpenAI API key is required. Set OPENAI_API_KEY environment variable or pass api_key parameter.")
        
        self.llm = OpenAI()

    def generate_response(self,query:str,vector_results: List[Dict[str,Any]]) -> str:
        # Generate response using OpenAI LLM based on user query and retrieved documents
        try:
            context = "\n\n".join([result['content'] for result in vector_results]) if vector_results else ""
            if not context:
                raise ValueError("No context available to generate a response.")

            prompt_template = PromptTemplate(
                input_variables=["context", "question"],
                template="""You are a helpful AI assistant. Use the following context to answer the question accurately and concisely.

                    Context:
                    {context}

                    Question: {question}

                    Answer: Provide a clear and informative answer based on the context above. If the context doesn't contain enough information to answer the question, say so."""
            )

            formatted_prompt = prompt_template.format(context=context, question=query)
            messages = [HumanMessage(content=formatted_prompt)]
            response = self.llm.invoke(messages)
            answer = response.choices[0].message.content.strip()
            print(f"✅ Response generated successfully!")
            return answer
        except Exception as e:
            print(f"❌ Error generating response from OpenAI LLM: {e}")
            raise e

    def generate_response_simple(self, query: str, context: str) -> str:
        """
        Simple response generation without complex prompting
        
        Args:
            query: User question
            context: Retrieved context
            
        Returns:
            Generated response
        """
        simple_prompt = f"""Based on this context: {context}

        Question: {query}

        Answer:"""
        
        try:
            messages = [HumanMessage(content=simple_prompt)]
            response = self.llm.invoke(messages)
            return response.content
        except Exception as e:
            return f"Error: {str(e)}"
```

## 🔍 BM25 Hybrid Search - Advanced Retrieval for Factual Data

### Why Hybrid Search?
Pure semantic similarity fails for **factual data retrieval** (phone numbers, company names, exact facts). Hybrid search combines **BM25 keyword matching** with **semantic similarity** to catch both exact matches and conceptual relationships.

**Problem Solved:** ✅ Retrieve phone numbers, company names, and structured data accurately
**Solution:** Use weighted combination of BM25 (keyword) + Semantic (concept) scores

```python
## 📊 Retrieval Strategy Guide - Weight Recommendations

guide_text = """
🎯 RETRIEVAL STRATEGY GUIDE - Why Hybrid Search Fixes Factual Data Loss

PROBLEM: Pure semantic similarity (original approach)
---------
❌ Phone numbers, company names, exact facts have LOW semantic similarity
❌ "Phone" and "+1-800-XXXX" are semantically different
❌ Named entities (Satender Singh, Axtria) may not cluster well
❌ Threshold filtering cuts off relevant results

✅ SOLUTION: Hybrid Retrieval (BM25 + Semantic)
--------
✅ BM25 catches EXACT KEYWORDS (phone, Axtria, Satender)
✅ Semantic similarity catches CONCEPT MATCHES (employment, contact)
✅ Combined score balances both approaches

📊 WEIGHT RECOMMENDATIONS:
-----------
For FACTUAL queries (phone, company, email, numbers):
  • Use: bm25_weight=0.6, semantic_weight=0.4
  • Example: "What is the phone number?" → keyword dominates

For CONCEPTUAL queries (summary, description, explain):
  • Use: bm25_weight=0.3, semantic_weight=0.7
  • Example: "Summarize Satender's profile" → semantic dominates

For MIXED queries (employment history, technical skills):
  • Use: bm25_weight=0.5, semantic_weight=0.5
  • Example: "Did person X work at company Y?" → balanced

USAGE:
------
hybrid_rag_retriever.retrieve(
    query="Your question",
    top_k=5,
    bm25_weight=0.6,      # ← Adjust based on query type
    semantic_weight=0.4
)
"""

print(guide_text)
```

```python
import re
from rank_bm25 import BM25Okapi

class HybridRAGRetriever:
    """
    Hybrid Retriever combining BM25 keyword matching + semantic similarity.
    This solves the problem of factual data retrieval where pure semantic search fails.
    """
    def __init__(self, vector_store, embedding_manager):
        self.vector_store = vector_store
        self.embedding_manager = embedding_manager
        self._token_pattern = re.compile(r"\w+", flags=re.UNICODE)

        # Load all documents from the vector store so BM25 token order matches IDs
        # Use only 'documents' and 'metadatas' to avoid requiring a data loader
        results_full = self.vector_store.collection.get(include=['documents', 'metadatas'])

        # Extract ids if returned by this client version
        doc_ids = []
        if 'ids' in results_full and results_full['ids'] and results_full['ids'][0]:
            doc_ids = results_full['ids'][0]

        self.doc_ids = doc_ids
        self.doc_texts = results_full.get('documents', [])
        self.metadatas = results_full.get('metadatas', [])

        if not self.doc_texts:
            raise ValueError("No documents found in the vector store collection for hybrid retrieval.")

        if not self.doc_ids:
            self.doc_ids = [f"doc_{i}" for i in range(len(self.doc_texts))]

        self.tokenized_docs = [self._tokenize_text(doc) for doc in self.doc_texts]
        self.bm25 = BM25Okapi(self.tokenized_docs)
        print(f"✅ Hybrid Retriever initialized with {len(self.doc_texts)} documents")

    def _tokenize_text(self, text: str) -> list:
        return self._token_pattern.findall(text.lower())

    def retrieve(self, query: str, top_k: int = 5, bm25_weight: float = 0.4, semantic_weight: float = 0.6) -> list:
        """
        Retrieve documents using a hybrid approach.

        Args:
            query: User query
            top_k: Number of results to return
            bm25_weight: Weight for BM25 scores (0-1)
            semantic_weight: Weight for semantic scores (0-1)

        Returns:
            List of retrieved documents with combined scores
        """
        try:
            print(f"🔄 Hybrid retrieval for query: '{query}'...")

            query_tokens = self._tokenize_text(query)
            print("  → Computing BM25 keyword scores...")
            bm25_scores = self.bm25.get_scores(query_tokens)

            bm25_max = max(bm25_scores) if len(bm25_scores) and max(bm25_scores) > 0 else 1
            bm25_scores_normalized = bm25_scores / bm25_max if bm25_max > 0 else bm25_scores

            print("  → Computing semantic similarity scores...")
            query_embedding = self.embedding_manager.generate_embeddings([query])[0]
            results = self.vector_store.collection.query(
                query_embeddings=query_embedding.tolist(),
                n_results=min(len(self.doc_ids), max(top_k * 5, 10)),
                include=['documents', 'metadatas', 'distances']
            )

            semantic_scores = {}
            if results.get('ids') and results['ids'][0]:
                for doc_id, distance in zip(results['ids'][0], results['distances'][0]):
                    semantic_scores[doc_id] = max(0.0, 1 - distance)

            print("  → Combining BM25 and semantic scores...")
            combined_results = []
            for idx, doc_id in enumerate(self.doc_ids):
                bm25_score = float(bm25_scores_normalized[idx]) if idx < len(bm25_scores_normalized) else 0.0
                semantic_score = float(semantic_scores.get(doc_id, 0.0))
                combined_score = (bm25_weight * bm25_score) + (semantic_weight * semantic_score)

                if combined_score > 0:
                    combined_results.append({
                        "id": doc_id,
                        "content": self.doc_texts[idx],
                        "metadata": self.metadatas[idx],
                        "bm25_score": bm25_score,
                        "semantic_score": semantic_score,
                        "combined_score": combined_score,
                    })

            sorted_results = sorted(combined_results, key=lambda x: x['combined_score'], reverse=True)
            for rank, result in enumerate(sorted_results[:top_k], 1):
                result['rank'] = rank

            print(f"✅ Retrieved {len(sorted_results[:top_k])} documents using hybrid search")
            return sorted_results[:top_k]

        except Exception as e:
            print(f"❌ Error in hybrid retrieval: {e}")
            raise e
```
